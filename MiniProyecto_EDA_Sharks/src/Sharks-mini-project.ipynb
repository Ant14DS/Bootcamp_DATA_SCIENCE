# Hipótesis 7. <a id="Hipo7"></a>
- Reunion Island es el lugar del mundo donde ocurren más ataques de tiburón en terminos relativos.

````python
'''
Para comprobar esta hipótesis voy a usar la media de ataques en un rango de varios años
y la voy a comparar con la media de población del país en ese mismo rango de años.

Usaré la siguiente tabla de paises y poblaciones:
https://databank.bancomundial.org

En esa fuente está la evolución de la población entre 1968 y 2020, pero no está Isla Reunión

Para Solucionarlo:
En este link está la tasa de crecimiento anual: https://countrymeters.info/es/Reunion

En google dan el dato de población de 2018, que es 833.944
con ese dato y las tasas de crecimiento anual, calculo las poblaciones de 2010 a 2017
'''
````

````python
#Veo en qué años tiene ataques REUNION ISLAND
reunion = years[years['Country']=='REUNION ISLAND'].groupby('Year')[['unos']].sum().sort_values('unos',ascending = False)
reunion.sort_values('Year',ascending = False)
````

````python
# Reunion tiene ataques cada año entre 2010 y 2017
# Elijo ese rango de años para validar la hipótesis
#Vuelvo a crear xpais, agrupado por país y año, entre 2010 y 2017
xpais = sharks[(sharks['Year'] >= 2010) & (sharks['Year'] <= 2017)]\
          .groupby(['Country','Year'])[['unos']].sum().sort_values('unos',ascending = False)
print(xpais.shape)
xpais.head(20)
````

````python
#Pivoto el df para ver los años como columnas
xpais = xpais.unstack()
xpais
````

````python
#Me quedo solo con los paises que tienen ataques en todos los años del rango 2010-2017.
xpais.dropna(axis = 0, thresh = 8, inplace = True)
````

````python
xpais
````

````python
#Calculo la media de ataques por pais en el rango de esos 8 años
xpais['attack_avg'] = xpais.apply(lambda x: round(x.mean(),2),axis = 1)
xpais = xpais.sort_values('attack_avg',ascending = False)
xpais
````

````python
#Pongo el pais como columna
xpais['country'] = xpais.index.get_level_values(0)
xpais
````

````python
#Para las poblaciones, cojo datos de bancomundial.org
#Del que he hecho un csv con las poblaciones entre 2010 y 2017,
#añadiendo manualmente Isla Reunion, de la forma que explico arriba
countries = pd.read_csv('../input/global_population.csv', encoding = 'latin1',sep = ";")
````

# Hipótesis 8. <a id="Hipo8"></a>
- En Sudáfrica hay más ataques letales a mujeres que en Australia.

````python
#Compruebo que no hay celdas duplicadas de Australia, por tener espacios en la celda y cosas así
sharks[sharks['Country'].str.contains('AUST')].groupby('Country').sum()
````

````python
#Selecciono solo mujeres con resultado Fatal_Y/N conocido
ausaf = sharks[(sharks['Sex']=='F')&(sharks['Fatal_Y/N']!='UNKNOWN')]
````

````python
#Me quedo con los registros de sudáfrica y australia
ausaf = ausaf[(ausaf['Country'] == 'SOUTH AFRICA')|(ausaf['Country'] == 'AUSTRALIA')]
ausaf.shape
````

````python
#agrupo por country y fatal_y/n
ausaf.groupby(['Country','Fatal_Y/N'])[['unos']].sum()
````

````python
plt.figure(figsize=(10,6))
colores = ['darkblue','lightblue']
sns.catplot('Country', data = ausaf, kind = 'count', hue = 'Fatal_Y/N', palette = colores)

sns.despine()
````

````python
ausaf2 = ausaf.groupby(['Country','Fatal_Y/N'])[['unos']].sum()
ausaf2 = ausaf2.unstack()
ausaf2
````

````python
ausaf2['country'] = ausaf2.index.get_level_values(0)
ausaf2
````

````python
ausaf2['total'] = ausaf2['unos']['Y'] + ausaf2['unos']['N']
ausaf2
````

````python
ausaf2['ratio'] = round(ausaf2['unos']['Y']/ausaf2['total']*100,2)
ausaf2
````
