# SEGUNDA HIPÓTESIS: Volar es cada vez más seguro<a id='H2'></a>

```python
''' 
En esta parte calculo LOS TOTALES de vuelos, personas trasportadas y poblaciones por país
Teniendo en cuenta que en las celdas que pone cero no cuenten.

Primero calculo el sumatorio de cada fila en una nueva columna
'''
#En df de vuelos/país/año
kbmfly_code['total_flight'] = kbmfly_code.loc[:,'1970':'2009'].apply(lambda x: x.sum(), axis = 1)

#En df de personas transportadas/país/año
kbmpeople_code['total_passengers'] = kbmpeople_code.loc[:,'1970':'2009'].apply(lambda x: x.sum(), axis = 1)

#En df de poblaición/país/año
kbmpopu_code['total_population'] = kbmpopu_code.loc[:,'1970':'2009'].apply(lambda x: x.sum(), axis = 1)
```

### En esta parte calculo los valores medios, aunque finalmente NO LO UTILIZO EN LA PRESENTACIÓN

```python
'''
Dado que algunos países no tienen datos en algunos años (el valor 0 significa 'desconocido'), hago 
una columna con la cantidad de ceros de cada país, para poder calcular una media basada en los
datos conocidos.
'''
#Creo columna con la cantidad de 0's de cada país, es decir, la cantidad de años en los que no 
#Hay dato conocido

#df vuelos/pais/año
kbmfly_code['count 0s'] = 0
for index, row in kbmfly_code.iterrows():
        kbmfly_code.loc[index,'count 0s'] = list(row).count(0)-1

#df personas transporatas/pais/año       
kbmpeople_code['count 0s'] = 0
for index, row in kbmpeople_code.iterrows():
        kbmpeople_code.loc[index,'count 0s'] = list(row).count(0)-1
        

#población mundial/pais/año        
kbmpopu_code['count 0s'] = 0
for index, row in kbmpopu_code.iterrows():
        kbmpopu_code.loc[index,'count 0s'] = list(row).count(0)-1
```

```python
#kbmfly_code.tail()
```

```python
#Como en el campo 'Total passengers' hay algún cero, en esas filas hay que restarle 1 al campo 'count 0's'

#en df personas transportadas/país/año
kbmpeople_code.loc[kbmpeople_code['total_passengers']==0,'count 0s'] = \
kbmpeople_code.loc[kbmpeople_code['total_passengers']==0,'count 0s']-1

#en df vuelos/pais año
kbmfly_code.loc[kbmfly_code['total_flights']==0,'count 0s'] = \
kbmfly_code.loc[kbmfly_code['total_flights']==0,'count 0s']-1

#en def población/país/año
kbmpopu_code.loc[kbmpopu_code['total_population']==0,'count 0s'] = \
kbmpopu_code.loc[kbmpopu_code['total_population']==0,'count 0s']-1
```

```python
#Compruebo
kbmfly_code.tail()
```

```python
'''
Una vez que tengo, en los 3 df's, el total de cada país y los años en los que hay datos,
puedo calcular el valor medio para cada país.

restando 40 menos la cantidad de 0's se obtiene el número entre el que hay que dividir
el total para hacer la media.
'''
#df vuelos/pais/año
kbmfly_code['avg_flights'] = 0

for index, row in kbmfly_code.iterrows():
    
    #Si el denominador es 0, para que no de error pongo 0 en el campo 'avg_flights'
    if kbmfly_code.loc[index,'count 0s'] == 0:
        
        kbmfly_code.loc[index,'avg_flights'] = \
        round(kbmfly_code.loc[index,'total_flights'] / 40, 2)
        
    #si el numerador es 0, para que no devuelva NaN pongo 0 en el campo 'avg_flights'
    elif kbmfly_code.loc[index,'total_flights'] == 0:
        
        kbmfly_code.loc[index,'avg_flights'] = 0
   
    #En el resto la media es 40 entre la cantidad de 0s
    else:
        
        kbmfly_code.loc[index,'avg_flights'] = \
        round(kbmfly_code.loc[index,'total_flights'] / (40 - kbmfly_code.loc[index,'count 0s']),2)

        
##df personas transporatas/pais/año       
kbmpeople_code['avg_passengers'] = 0

for index, row in kbmpeople_code.iterrows():
    
    if kbmpeople_code.loc[index,'count 0s'] == 0:
        
        kbmpeople_code.loc[index,'avg_passengers'] = \
        round(kbmpeople_code.loc[index,'total_passengers'] / 40,2)
        
    elif kbmpeople_code.loc[index,'total_passengers'] == 0:
        
        kbmpeople_code.loc[index,'avg_passengers'] = 0
   
    else:
        
        kbmpeople_code.loc[index,'avg_passengers'] = \
        round(kbmpeople_code.loc[index,'total_passengers'] / (40 - kbmpeople_code.loc[index,'count 0s']),2)

        
##df población mundial/pais/año        
kbmpopu_code['avg_population'] = 0

for index, row in kbmpopu_code.iterrows():
    
    if kbmpopu_code.loc[index,'count 0s'] == 0:
        
        kbmpopu_code.loc[index,'avg_population'] = \
        round(kbmpopu_code.loc[index,'total_population'] / 40,2)
        
    elif kbmpopu_code.loc[index,'total_population'] == 0:
        
        kbmpopu_code.loc[index,'avg_population'] = 0
   
    else:
        
        kbmpopu_code.loc[index,'avg_population'] = \
        round(kbmpopu_code.loc[index,'total_population'] / (40 - kbmpopu_code.loc[index,'count 0s']),2)
```

### CALCULO LOS SUMATORIOS QUE SÍ USO EN LA PRESENTACIÓN

```python
'''
Una vez que tengo los valores medios por país, obtengo los valores totales por año, añado una fila con el 
sumatorio.
'''
#En df de vuelos/país/año
kbmfly_code.loc[217,:] = kbmfly_code.loc[:,'1970':'2009'].apply(lambda x: x.sum(), axis = 0)
kbmfly_code.fillna('totalfly',inplace = True)

#En df de personas transportadas/país/año
kbmpeople_code.loc[217,:] = kbmpeople_code.loc[:,'1970':'2009'].apply(lambda x: x.sum(), axis = 0)
kbmpeople_code.fillna('totalpeople',inplace = True)

#En df de poblaición/país/año
kbmpopu_code.loc[215,:] = kbmpopu_code.loc[:,'1970':'2009'].apply(lambda x: x.sum(), axis = 0)
kbmpopu_code.fillna('totalpopu',inplace = True)

kbmfly_code.iloc[-1]
```

```python
#RECTIFICACIÓN: Al hacer el fillna, he convertido algunas columnas numéricas a objeto, lo rectifico:
#Primero el valor de la última fila lo cambio a 0, después convierto a numéricas las columnas

kbmfly_code.loc[217,['total_flights','avg_flights','count 0s']] = 0
kbmfly_code['total_flights'] = kbmfly_code['total_flights'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')
kbmfly_code['count 0s'] = kbmfly_code['count 0s'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')
kbmfly_code['avg_flights'] = kbmfly_code['avg_flights'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')

kbmpeople_code.loc[217,['total_passengers','avg_passengers','count 0s']] = 0
kbmpeople_code['total_passengers'] = kbmpeople_code['total_passengers'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')
kbmpeople_code['count 0s'] = kbmpeople_code['count 0s'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')
kbmpeople_code['avg_passengers'] = kbmpeople_code['avg_passengers'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')

kbmpopu_code.loc[215,['total_population','avg_population','count 0s']] = 0
kbmpopu_code['total_population'] = kbmpopu_code['total_population'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')
kbmpopu_code['count 0s'] = kbmpopu_code['count 0s'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')
kbmpopu_code['avg_population'] = kbmpopu_code['avg_population'].apply(pd.to_numeric,errors = 'ignore',downcast = 'float')
```

```python
'''
df agrupado por regiones, total de vuelos
'''
total_vuelos = kbmfly_code.groupby('region')[['total_flights']].sum().sort_values('total_flights',ascending = False)
total_vuelos
```

```python
#Total vuelos por regiones BARRAS. Compañías registradas en países pertenecientes a las regiones
#DESCARTO LAS BARRAS Y ME QUEDO CON EL TREEMAP

fig, ax = plt.subplots(figsize=(12, 8))

plt.bar(total_vuelos.index,'total_flights',data=total_vuelos,color='salmon',linewidth=0.5)
#sns.barplot(x=listmonth,y="count", data=bymonth,color="salmon")

ax.set_xlabel('Region',fontsize=20)
ax.set_ylabel('Nº Vuelos en Centenas de Millones',fontsize=20)

#'center', 'top', 'bottom', 'baseline', 'center_baseline
#ax.text(4,0,s = 'Nº de vuelos globales de compañías registradas\nen países de las regiones ',va = 'top', fontsize=5);

ax.tick_params(axis='x', labelrotation=75 )
ax.set_title('Promedio de vuelos/año entre 1970-2009 de compañías registradas',loc='Center',fontsize=20);
plt.tight_layout();
```

```python
#Total vuelos por regiones TREEMAP. Compañías registradas en países de las regiones mostradas

cmap2 = mpl.cm.Pastel1#tab20c
mini=total_vuelos['total_flights'].min()
maxi=total_vuelos['total_flights'].max()
norm = mpl.colors.Normalize(vmin=mini, vmax=maxi)
colors2 = [cmap2(norm(value)) for value in total_vuelos['total_flights']]

fig, ax = plt.subplots(figsize=(12, 8))

#plt.figure(figsize=(12,8))

squarify.plot(sizes=total_vuelos.values,
             label=total_vuelos.index,
             alpha=.6, color = colors2,bar_kwargs=dict(linewidth=0.3, edgecolor="#111111"))
plt.axis('off')
plt.tight_layout();
```

```python
#Al comparar las víctimas del set de Kaggle con el de Airfleets, se comprueba que 
#La difernecia entre accidentes anuales es muy grande
plt.figure(figsize=(10,4))

plt.style.use('fivethirtyeight')

plt.plot(af_byear.index,'fatalities',data=af_byear,color='seagreen',linewidth=0.5)
plt.plot(k_byear.index,'Fatalities',data=k_byear,color='blue',linewidth=1)

plt.xlabel('Año',fontsize=15)
plt.ylabel('Cantidad Personas',fontsize=15)
plt.title('Comparativa víctimas Kaggle/Airfleets 1970-2009',loc='Center',fontsize=20)

plt.xlim(1970,2009.1)
plt.legend()
plt.tight_layout();
```

```python
'''
Comparando la cantidad de accidentes por año de Kaggle con Airfleets, la diferencia es muy grande,
la razón más probable es que en Kaggle estén todos los accidentes de todos los tipos de aviones
y en airfleets solamente los comerciales de las grandes líneas aéreas.

Hago un df confrontando los accidentes/año de ada dataset
'''
prb1 = air[air['year']>=1970].groupby('year')[['unos']].count()
prb2 = airfleets.groupby('year')[['aircraft']].count()
#prb3 = pd.concat([prb1,prb2],ignore_index = True)
prb3 = prb1.join(prb2)
prb3.rename(columns = {'unos':'Kaggle','aircraft':'airfleets'},inplace = True)
prb3.head()
```
